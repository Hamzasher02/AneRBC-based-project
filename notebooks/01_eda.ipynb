# Task 1 & 5: Exploratory Data Analysis (EDA) on AneRBC Dataset

This notebook verifies the dataset counts, class distribution, and split distribution, and visualises sample images from the healthy and anaemic classes.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image

splits_dir = Path("../data/processed")
train_df = pd.read_csv(splits_dir / "train.csv")
val_df = pd.read_csv(splits_dir / "val.csv")
test_df = pd.read_csv(splits_dir / "test.csv")

print(f"Train set size      : {len(train_df)}")
print(f"Validation set size : {len(val_df)}")
print(f"Test set size       : {len(test_df)}")
print(f"Total dataset size  : {len(train_df) + len(val_df) + len(test_df)}")
```

### Class Distribution check

Verify that the split is perfectly stratified and check the class balance of the dataset.

```python
print("--- Training Class Distribution ---")
print(train_df["class_name"].value_counts())

print("\n--- Validation Class Distribution ---")
print(val_df["class_name"].value_counts())

print("\n--- Test Class Distribution ---")
print(test_df["class_name"].value_counts())
```

### Plot Class Distribution

Visualise class balanced split.

```python
all_df = pd.concat([train_df, val_df, test_df])
all_df["class_name"].value_counts().plot(kind="bar", color=["skyblue", "salmon"])
plt.title("Class Distribution in AneRBC Dataset")
plt.xlabel("Class")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

### Display Sample Images

Load and display a healthy and an anaemic red blood cell (RBC) sample image.

```python
healthy_sample = Path("../data/processed/healthy/001_h.jpg")
anaemic_sample = Path("../data/processed/anaemic/001_a.jpg")

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(Image.open(healthy_sample))
axes[0].set_title("Healthy RBC Sample (001_h.jpg)")
axes[0].axis("off")

axes[1].imshow(Image.open(anaemic_sample))
axes[1].set_title("Anaemic RBC Sample (001_a.jpg)")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```
