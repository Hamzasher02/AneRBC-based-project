# Task 3: Transfer Learning Comparison Experiments

This notebook documents the transfer learning experiments with three pretrained CNN architectures: `mobilenet_v2`, `squeezenet1_0`, and `resnet18`. All models are fine-tuned as feature extractors with frozen convolutional backbones.

### Pretrained Architecture Details

| Pretrained Model | PyTorch Backbone | Replaced Classifier Layer | Frozen Backbone Layers | Trainable Top Layers |
| :--- | :--- | :--- | :--- | :--- |
| **MobileNetV2** | `mobilenet_v2` | `classifier[1]` (Linear) | Features 0 to 18 | `classifier.1.weight`, `classifier.1.bias` |
| **SqueezeNet 1.0** | `squeezenet1_0` | `classifier[1]` (Conv2d) | Features 0 to 12 | `classifier.1.weight`, `classifier.1.bias` |
| **ResNet18** | `resnet18` | `fc` (Linear) | Layers 1 to 4 | `fc.weight`, `fc.bias` |

The convolutional backbones are completely frozen using `requires_grad = False` to prevent overfitting on the relatively small medical image dataset.

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image

reports_dir = Path("../outputs/reports")
val_csv = reports_dir / "transfer_validation_summary.csv"
test_csv = reports_dir / "transfer_test_summary.csv"

if val_csv.exists():
    print("--- Pretrained Validation Performance Summary ---")
    print(pd.read_csv(val_csv).to_string(index=False))
else:
    print("transfer_validation_summary.csv not found!")
```

### Pretrained Models Test performance Summary

The checkpoints were evaluated on the independent test set (150 images, stratified).

```python
if test_csv.exists():
    print("--- Pretrained Test Performance Summary ---")
    print(pd.read_csv(test_csv).to_string(index=False))
else:
    print("transfer_test_summary.csv not found!")
```

### Best Pretrained Model visualisations

Load and display validation curve and confusion matrix for the best transfer learning model (`mobilenet_v2`).

```python
curve_path = Path("../outputs/figures/learning_curve_mobilenet_v2.png")
cm_path = Path("../outputs/figures/confusion_matrix_mobilenet_v2.png")

if curve_path.exists():
    plt.figure(figsize=(10, 5))
    plt.imshow(Image.open(curve_path))
    plt.title("Learning Curves - MobileNetV2")
    plt.axis("off")
    plt.show()

if cm_path.exists():
    plt.figure(figsize=(7, 7))
    plt.imshow(Image.open(cm_path))
    plt.title("Confusion Matrix - MobileNetV2")
    plt.axis("off")
    plt.show()
```
