# Task 2: Custom CNN Architectures & Training Experiments

This notebook documents the custom CNN models with 3, 4, and 5 convolutional layers, their configurations, validation curves, and test performance. The best custom model is selected and analyzed.

### Custom CNN Architectures Summary

| Model Name | Conv Layers | Filters per layer | Dropout Rate | Classifier Neurons (FC) |
| :--- | :---: | :--- | :---: | :--- |
| `custom_cnn_3` | 3 | `[32, 64, 128]` | 0.25 | `[128, 256, 2]` |
| `custom_cnn_4` | 4 | `[32, 64, 128, 256]` | 0.30 | `[256, 512, 2]` |
| `custom_cnn_5` | 5 | `[32, 64, 128, 256, 512]` | 0.40 | `[512, 256, 2]` |

All custom models use standard `ConvBlock` containing: `nn.Conv2d` -> `nn.BatchNorm2d` -> `nn.ReLU` -> `nn.MaxPool2d` (2x2), followed by a Global Adaptive Average Pooling layer.

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image

reports_dir = Path("../outputs/reports")
val_csv = reports_dir / "custom_cnn_validation_summary.csv"
test_csv = reports_dir / "custom_cnn_test_summary.csv"

if val_csv.exists():
    print("--- Validation Summary CSV ---")
    print(pd.read_csv(val_csv).to_string(index=False))
else:
    print("custom_cnn_validation_summary.csv not found!")
```

### Custom CNN Test Results Comparison

The checkpoints were evaluated on the independent test set.

```python
if test_csv.exists():
    print("--- Test Performance Summary ---")
    print(pd.read_csv(test_csv).to_string(index=False))
else:
    print("custom_cnn_test_summary.csv not found!")
```

### Best Custom CNN Model Visualisations

Load and display the learning curves and test set confusion matrix for `custom_cnn_4`, which achieved the best test metrics.

```python
curve_path = Path("../outputs/figures/learning_curve_custom_cnn_4.png")
cm_path = Path("../outputs/figures/confusion_matrix_custom_cnn_4.png")

if curve_path.exists():
    plt.figure(figsize=(10, 5))
    plt.imshow(Image.open(curve_path))
    plt.title("Learning Curves - Custom CNN 4")
    plt.axis("off")
    plt.show()

if cm_path.exists():
    plt.figure(figsize=(7, 7))
    plt.imshow(Image.open(cm_path))
    plt.title("Confusion Matrix - Custom CNN 4")
    plt.axis("off")
    plt.show()
```
