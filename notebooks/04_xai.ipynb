# Task 4: Explainable AI (XAI) Visualisations

This notebook covers the application of Explainable AI (XAI) algorithms on the best custom model (`custom_cnn_4`) and the best pretrained model (`mobilenet_v2`) using Captum. We display and compare saliency maps generated via **Grad-CAM** and **Integrated Gradients**.

### XAI Explanations Overview

- **Grad-CAM**: Visualises regional attribution by mapping gradients flowing into the final convolutional layer of the model. Shows coarse heatmaps indicating the general regions driving predictions.
- **Integrated Gradients**: Computes attribution by integrating gradients along path from a neutral baseline (black image) to the input image. Shows pixel-level fine attribution features.

```python
import json
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image

summary_path = Path("../outputs/reports/xai_summary.json")
if summary_path.exists():
    with open(summary_path, "r", encoding="utf-8") as f:
        summary_data = json.load(f)
    df = pd.DataFrame(summary_data)
    print("--- XAI Summary JSON ---")
    print(df.to_string(index=False))
else:
    print("xai_summary.json not found!")
```

### Visualise Saliency Maps

Compare Grad-CAM and Integrated Gradients maps on the same anaemic RBC sample image (`165_a.jpg`).

```python
xai_dir = Path("../outputs/figures/xai")
models = ["custom_cnn_4", "mobilenet_v2"]
methods = ["gradcam", "integrated_gradients"]

for model in models:
    print(f"\n=== Saliency maps for {model} ===")
    fig, axes = plt.subplots(1, len(methods), figsize=(12, 6))
    for i, method in enumerate(methods):
        path = xai_dir / f"{method}_{model}_165_a.png"
        if path.exists():
            axes[i].imshow(Image.open(path))
            axes[i].set_title(f"{method.replace('_', ' ').title()}")
            axes[i].axis("off")
        else:
            axes[i].text(0.5, 0.5, f"{path.name}\nnot found", ha="center", va="center")
            axes[i].axis("off")
    plt.tight_layout()
    plt.show()
```

### Interpretation & Qualitative Critique

1. **Attribution Focus**: Both models focus attributions on the central and cell boundary regions of the red blood cells, which is clinically meaningful as cell shape, pallor, and boundaries differ in anaemic cells.
2. **Grad-CAM coarse heatmap**: Grad-CAM highlights general regions, showing high activations on the cell itself.
3. **Integrated Gradients pixel attribution**: Integrated Gradients provides high-frequency edge attributions, highlighting shapes and texture changes.
4. **Pretrained model (MobileNetV2)**: Shows more structured, localized attributions corresponding closely to the cells compared to the custom CNN model, aligning with its higher overall performance.
